```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Gradient_Descent(theta, alpha, epsilon, num_iters, f, grad_f):
    x_history = [theta]
    for i in range(num_iters):
        theta = theta - alpha * grad_f(theta)
        x_history.append(theta)
        print("epoch: {}, x new: {}".format(i, theta))
        if f(theta) < epsilon:
            break
    return x_history
```

```python
def f(x):
    return x**2 + x + 1

def grad_f(x):
    return 2 * x + 1

theta = 15
alpha = 0.1
epsilon = 0.0001
num_iters = 50

his_x = Gradient_Descent(theta, alpha, epsilon, num_iters, f, grad_f)
```

```python
plt.plot(np.linspace(-15, 15, 100), f(np.linspace(-15, 15, 100)))
plt.plot(his_x, f(np.array(his_x)), 'ro')
plt.annotate('local min', xy=(his_x[-1], f(his_x[-1])), xytext=(his_x[-1], f(his_x[-1])+100), arrowprops=dict(facecolor='black', shrink=0.05))
```
