```python
def Gradient_Descent(theta, alpha, epsilon, num_iters, f, grad_f):
    """
    Performs gradient descent to learn theta
    by taking num_items gradient steps with learning
    rate alpha
    
    Parameters
    ----------
    X : array_like
        The input dataset of shape (m x n+1)
        m is the number of examples
        n is the number of features
    
    y : arra_like
        Value at given features
    
    theta : array_like
        Initial values for the linear regression parameters
    
    alpha : float
        The learning rate
    
    num_iters : int
        Number of iterations to update the parameters
    
    epislon : float
        Error threshold
        
    f : function

    grad_f : function
        Derivative of the function f
    
    Returns
    -------
    list of theta history
    """
    theta_history = [theta]
    for i in range(num_iters):
        theta = theta - alpha * grad_f(theta)
        theta_history.append(theta)
        if f(theta) < epsilon:
            break
    return theta_history
```

```python
def f(x):
    return (2*x+1)**2

def grad_f(x):
    return 4*(2*x+1)

theta = -10
alpha = 0.01
epsilon = 0.0001
num_iters = 1000

list_theta = Gradient_Descent(theta, alpha, epsilon, num_iters, f, grad_f)
```
